```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

## KFOLD

```python
methods = ["PCA & SVM", "Modular PCA & LR", "Kernel PCA & SVM", "Sparse PCA & SVM", "Randomized PCA & SVM", "Incremental PCA & SVM"]
accuracy = [0.98, 0.97, 0.98, 0.98, 0.98, 0.98]
f1_score = [0.971, 0.9565, 0.9718, 0.9718, 0.9718, 0.9718]

# Vẽ biểu đồ dòng
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(methods, accuracy, marker='s', label='Accuracy', color='#922a56')
ax.plot(methods, f1_score, marker='s', label='F1 Score', color='#031c65')

ax.set_xticklabels(methods, rotation=35, ha='center')
ax.set_yticklabels([])

ax.spines[['left','right','top']].set_visible(False)
ax.tick_params(axis='both', which='both', top=False, bottom=False, left=False, right=False)

ax.spines[['top', 'right']].set_visible(False)
ax.spines[['left', 'bottom']].set_color('#DCDCDC')

for i, (acc, f1, method) in enumerate(zip(accuracy, f1_score, methods)):
    if method != "Modular PCA & LR":
        ax.annotate(f'{round(acc*100, 3)}%', (method, acc), textcoords="offset points", xytext=(0,10), ha='center')
        ax.annotate(f'{round(f1*100, 3)}%', (method, f1), textcoords="offset points", xytext=(0,10), ha='center')

min_acc_index = accuracy.index(min(accuracy))
ax.annotate(f'{round(min(accuracy)*100, 3)}%', 
            xy=(methods[min_acc_index], accuracy[min_acc_index]), 
            xytext=(0, -17.5), 
            textcoords='offset points', 
            ha='center', 
            color='black')

# Annotate lowest f1-score
min_f1_index = f1_score.index(min(f1_score))
ax.annotate(f"{round(min(f1_score)*100, 3)}% ", 
            xy=(methods[min_f1_index], f1_score[min_f1_index]), 
            xytext=(0, -17.5), 
            textcoords='offset points', 
            ha='center', 
            color='black')
    
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 0.22))

kws = dict(va='bottom', size=18, weight='bold')
fig.text(0.016, 1.2, 'Top-performing models based on Accuracy & F1-score', **kws)
fig.text(0.30185, 1.125, 'from k-fold cross-validation', ha='right', va='bottom', size=15, alpha=0.8)
ax.set_yticks([0.95, 0.975, 0.98])

plt.tight_layout()
plt.show()
```
